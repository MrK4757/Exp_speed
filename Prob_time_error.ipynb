```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.size'] = 11
plt.rcParams['figure.dpi'] = 100
```

```python
df = pd.read_excel("test_result.xlsx", sheet_name='Sheet2')
df.head()
```

```python
rgbm = np.abs(df['Error_rgbm'], dtype=np.float32)
ffnn = np.abs(df['Error_Tfnn'], dtype=np.float32)
dbm = np.abs(df['Error_dbm'], dtype=np.float32)

print(rgbm[:7])
```

```python
rgbm = (df['Error_rgbm'])
ffnn = (df['Error_Tfnn'])
dbm =  (df['Error_dbm'])
```

```python
plt.figure(figsize=(6,7))

plt.subplot(3,1,1)
sns.histplot(data = df, x = rgbm, stat = "probability", label = 'RGBM', kde=True)
plt.xlabel('Transit time error (hrs)')
plt.annotate(f"(a)", xy=(0.5, 0.95), xycoords='axes fraction', fontsize=11, ha = 'center', va= 'top')
plt.ylim(0, .5)
plt.xlim(0, 40)
plt.legend(frameon=False)


plt.subplot(3,1,2)
sns.histplot(data = df, x = ffnn, stat = "probability", label = 'FFNN', kde=True, color='blue')
plt.xlabel('Transit time error (hrs)')
plt.annotate(f"(b)", xy=(0.5, 0.95), xycoords='axes fraction', fontsize=11, ha='center', va = 'top')
plt.ylim(0, .5)
plt.xlim(0, 40)
plt.legend(frameon=False)

plt.subplot(3,1,3)
sns.histplot(data = df, x = dbm, stat = "probability", label = 'DBM', kde=True, color ='green')
plt.annotate(f"(c)", xy=(0.5, 0.95), xycoords='axes fraction', fontsize=11, ha = 'center', va= 'top')
plt.xlabel('Transit time error (hrs)')
plt.ylim(0, .5)
plt.xlim(0, 40)
plt.legend(frameon=False)

plt.tight_layout()
```
