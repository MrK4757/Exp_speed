```python
import pandas as pd


df = pd.read_excel("/media/mrk/Data/E-JUST/publication/Exp_speed/Exp_speed_new/test_set.xlsx")
df.head()
```

```python
import random
random.seed(1234)

t1 = []
t2 = []
t3 = []
t3 = []
t4 = []
v1 = []
v2 = []
v3 = []
v4 = []


# estimating the propagation speed and transit time based on Vlat and Vfront
for i in range(len(df['Vp'])):
    vlat = round(df['Vlat'].values[i], 0)* 1.0460
    vfrt = round(df['Vfrt'].values[i], 0)* 4.2189
    k = -97.0919
    vp = vlat + vfrt + k
    v1.append(round(vp, 0))
    t1.append(round(150000000/(vp*3600), 2))

# estimating the propagation speed and transit time based on Vlat Vfront and Vsw
for j in range(len(df['Vp'])):
    vlat = round(df['Vlat'].values[j], 0)* 0.7815
    vfrt = round(df['Vfrt'].values[j], 0)* 3.3242
    vsw = round(df['Vsw'].values[j], 0)* 0.5390
    k = -205.379
    vp = vlat + vfrt + vsw + k
    v2.append(round(vp, 0))
    t2.append(round(150000000/(vp*3600), 2))

# estimating the propagation speed and transit time based on Vlat Vfront and AW
for k in range(len(df['Vp'])):
    vlat = round(df['Vlat'].values[k], 0)* 0.8666
    vfrt = round(df['Vfrt'].values[k], 0)* 3.4519
    aw =   df['2aw'].values[k] * 2.1126
    k = -191.0689
    vp = vlat + vfrt + aw + k
    v3.append(round(vp, 0))
    t3.append(round(150000000/(vp*3600), 2))

# estimating the propagation speed and transit time based on Vlat Vfront AW and Vsw

for m in range(len(df['Vp'])):
    vlat = round(df['Vlat'].values[m], 0)* 0.7796
    vfrt = round(df['Vfrt'].values[m], 0)* 3.2320
    vsw = df['Vsw'].values[m]* 0.5618
    aw =  df['2aw'].values[m]* -0.1087
    k = -205.1152
    vp = vlat + vfrt + aw + vsw + k
    v4.append(round(vp, 0))
    t4.append(round(150000000/(vp*3600), 2))
```

```python
print(v1[:6])
print(t1[:6])
print()
print(v2[:6])
print(t2[:6])
print()
print(v3[:6])
print(t3[:6])
print()
print(v4[:6])
print(t4[:6])
```

```python
df['V1'] = v1
df['V2'] = v2
df['V3'] = v3
df['V4'] = v4
df['T1'] = t1
df['T2'] = t2
df['T3'] = t3
df['T4'] = t4

df.to_excel("/media/mrk/Data/E-JUST/publication/Exp_speed/Exp_speed_new/test_set.xlsx", index= False)
```
